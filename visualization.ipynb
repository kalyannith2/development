```python
# prompt: how to show bar value on the bar in chart

import matplotlib.pyplot as plt
# showing bar value on the bar in chart
cyl=mtcars.cyl.value_counts()
bars = plt.bar(cyl.index,cyl)
plt.ylabel('frequency')

# Add the values on top of the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), va='bottom', ha='center') # va: vertical alignment, ha: horizontal alignment

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
dir(sns)
```

```python
tips=sns.load_dataset('tips')
```

```python
tips
```

```python
tips.dtypes
```

```python
#strip plot

sns.stripplot(tips['tip'])
plt.ylabel('tip($)')
plt.show()
```

```python
sns.stripplot(tips['tip'],jitter=False)  #jitter:one line
plt.ylabel('tip($)')
plt.show()
```

```python
sns.stripplot(x='day',y='tip',data=tips,size=4,jitter=True)
plt.ylabel('tip($)')
plt.show()
```

```python
#swarm plot

sns.swarmplot(x='day',y='total_bill',data=tips)
plt.ylabel('total_bill($)')
plt.show()

```

```python
sns.swarmplot(x='day',y='total_bill',data=tips,hue='sex')   #hue:divide indivisual
plt.ylabel('total_bill($)')
plt.show()
```

```python
import warnings                       #avoid the warnigs
warnings.filterwarnings('ignore')
```

```python
#changing orientation
sns.swarmplot(x='total_bill',y='day',data=tips,hue='smoker',orient='h')
plt.ylabel('total_bill($)')
plt.show()
```

```python
help(sns.swarmplot)
```

```python
# box plot and violin plot

#plt.subplot(1,2,1)
sns.boxplot(x='day',y='total_bill',data=tips,hue='time')
plt.ylabel('total_bill($)')
plt.show()
```

```python
sns.boxplot(tips['total_bill'])
```

```python
sns.boxplot(x='time',y='total_bill',data=tips,hue='day',legend='auto')
plt.ylabel('total_bill($)')
plt.show()
```

```python
sns.violinplot(x='day',y='tip',data=tips,hue='sex')
plt.ylabel('tip($)')
plt.show()
```

```python
sns.violinplot(x='day',y='tip',data=tips,hue='sex',inner=None)
plt.ylabel('tip($)')
plt.show()
```

```python
plt.subplot(2,1,1)
sns.violinplot(x='day',y='tip',data=tips,inner=None)
plt.ylabel('tip($)')
plt.show()
plt.subplot(2,1,2)  #(2,1,2)
sns.boxplot(x='time',y='tip',data=tips)
plt.ylabel('tip($)')
plt.show()
```

```python
#combining plots

sns.violinplot(x='day',y='tip',data=tips,inner=None,color='lightgray')
#sns.stripplot(x='day',y='tip',data=tips,size=4,jitter=True)
sns.boxplot(x='day',y='tip',data=tips,hue='day')
plt.ylabel('tip($)')
plt.show()

```

# titanic data visulation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
sns.load_dataset('titanic')
```

```python
data=pd.read_csv('train.csv')
data.head(10)
```

```python
data.groupby('Pclass')['Survived'].count()
```

```python
f,ax=plt.subplots(figsize=(11,5))
sns.boxplot(x='Survived',y='Age',data=data)
plt.show()
```

```python
f,ax=plt.subplots(figsize=(10,5))
sns.barplot(x='Sex',y='Survived',data=data)
plt.ylabel('Survived persentage')
plt.show()
```

```python
f,ax=plt.subplots(figsize=(11,5))
sns.boxplot(x='Sex',y='Age',hue='Survived',data=data)
```

```python
sns.barplot(x='Pclass',y='Survived',data=data)
```

```python
sns.barplot(x='Pclass',y='Survived',hue='Sex',data=data)
```

```python
sns.barplot(x='SibSp',y='Survived',data=data)
```

```python
survived=data.loc[data['Survived']==1,'Age'].dropna()   #dropna:drop or dellet the nall values
sns.distplot(survived)
plt.title('Survived')
plt.show()
```

```python
not_survived=data.loc[data['Survived']==0,'Age'].dropna()
sns.distplot(not_survived)
plt.title('Not Survived')
plt.show()
```

```python
sns.distplot(data['Age'])
```

```python
sns.pairplot(data)
```

```python
# pclss vs survived
grid=sns.FacetGrid(data,col='Survived',row='Pclass',aspect=1)
grid.map(plt.hist,'Age',alpha=0.9,bins=20,edgecolor='red')
grid.add_legend()
```

```python
#data=
```

```python
sns.distplot(a=data[data['Embarked']=='C']['Survived'],bins=3,kde=False)
plt.title('Cherbourg')
plt.show()
sns.distplot(a=data[data['Embarked']=='Q']['Survived'],bins=3,kde=False)
plt.title('Queenstown')
plt.show()
sns.distplot(a=data[data['Embarked']=='S']['Survived'],bins=3,kde=False)
plt.title('S')
```

```python
figbi, axesbi = plt.subplots(2, 4, figsize=(16, 10))
data.groupby('Pclass')['Survived'].mean().plot(kind='barh',ax=axesbi[0,0],xlim=[0,1])
data.groupby('SibSp')['Survived'].mean().plot(kind='barh',ax=axesbi[0,1],xlim=[0,1])
data.groupby('Parch')['Survived'].mean().plot(kind='barh',ax=axesbi[0,2],xlim=[0,1])
data.groupby('Sex')['Survived'].mean().plot(kind='barh',ax=axesbi[0,3],xlim=[0,1])
data.groupby('Embarked')['Survived'].mean().plot(kind='barh',ax=axesbi[1,0],xlim=[0,1])
sns.boxplot(x="Survived", y="Age", data=data,ax=axesbi[1,1])
sns.boxplot(x="Survived", y="Fare", data=data,ax=axesbi[1,2])
sns.distplot(not_survived,ax=axesbi[1,3])
```

```python
figbi, axesbi = plt.subplots(2, 2, figsize=(16, 10))
data.groupby('Pclass')['Survived'].mean().plot(kind='barh',ax=axesbi[0,0],xlim=[0,1])
data.groupby('SibSp')['Survived'].mean().plot(kind='barh',ax=axesbi[0,1],xlim=[0,1])
sns.boxplot(x="Survived", y="Age", data=data,ax=axesbi[1,0])
sns.boxplot(x="Survived", y="Fare", data=data,ax=axesbi[1,1])
```

```python
figbi, axesbi = plt.subplots(2, 4, figsize=(16, 10))
data.groupby('Pclass')['Survived'].mean().plot(kind='barh',ax=axesbi[0,0],xlim=[0,1])
data.groupby('SibSp')['Survived'].mean().plot(kind='pie',ax=axesbi[0,1],autopct='%.2f%%')
data.groupby('Parch')['Survived'].mean().plot(kind='kde',ax=axesbi[0,2],xlim=[0,1])
data.groupby('Sex')['Survived'].mean().plot(kind='barh',ax=axesbi[0,3],xlim=[0,1])
data.groupby('Embarked')['Survived'].mean().plot(kind='barh',ax=axesbi[1,0],xlim=[0,1])
sns.boxplot(x="Survived", y="Age", data=data,ax=axesbi[1,1])
sns.boxplot(x="Survived", y="Fare", data=data,ax=axesbi[1,2])
sns.distplot(not_survived,ax=axesbi[1,3])
```
